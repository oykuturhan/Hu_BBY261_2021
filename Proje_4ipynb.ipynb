```python
import tensorflow as tf


import numpy as np
import logging
logger = tf.get_logger()
logger.setLevel(logging.ERROR)

Baska=  int(input("Tekrar sayısını yazınnız: "))
Deger= int(input("Km dğerini giriniz: "))
Sonuc=Deger/1.609344


Kilometre  = np.array([10, 50,  30,  80,  45,  55,  70 ],  dtype=float)
Mil = np.array([6.21,  31.6,18.64,  49.70,  27.96,  34.17, 43.49],  dtype=float)
for i,c in enumerate(Kilometre):
  print("{} degrees Kilometre = {} degrees mil".format(c, Mil[i]))
l0 = tf.keras.layers.Dense(units=1, input_shape=[1])
model = tf.keras.Sequential([l0])
model = tf.keras.Sequential([
  tf.keras.layers.Dense(units=1, input_shape=[1])])
model.compile(loss='mean_squared_error',
              optimizer=tf.keras.optimizers.Adam(0.1))


history = model.fit(Kilometre, Mil, epochs=Baska, verbose=False)
#print("Finished training the model")
import matplotlib.pyplot as plt
plt.xlabel('Epoch Number')
plt.ylabel("Loss Magnitude")
plt.plot(history.history['loss'])
TahminiSonuc= (model.predict([Deger]))
print("Tahmin Etiiği: ", TahminiSonuc)
print("Gerçek Sonuç:", Sonuc)

```
